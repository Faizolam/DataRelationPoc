```python
Intdf1=[]
Vardf1=[]
Datedf1=[]
for col,dtype in df1.dtypes.iteritems():
#     print(col,dtype)
    if dtype=='int64':
        Intdf1.append(col)
    elif dtype=='object':
        Vardf1.append(col)
    elif dtype=='datetime64[ns]':
         Datedf1.append(col)
    
print(colored("*******************************************COLUMNS FROM TABLE basic_data******************************************************",'green'))        
print(colored("TYPE INT:",'red'))
print(colored(f'{Intdf1}\n','blue'))
print(colored("TYPE VARCHAR:",'red'))
print(colored(f' {Vardf1}','blue'))

Intdf2=[]
Vardf2=[]
Datedf2=[]
for col1,dtype1 in df2.dtypes.iteritems():
#     print(col,dtype)
    if dtype1=='int64':
        Intdf2.append(col1)
    elif dtype1=='object':
        Vardf2.append(col1)
    elif dtype=='datetime64[ns]':
         Datedf1.append(col)
print(colored("*****************************************COLUMNS FROM TABLE plant_data******************************************************",'green'))                    
print(colored("TYPE INT:",'red'))
print(colored(f'{Intdf2}\n','blue'))
print(colored("TYPE VARCHAR:",'red'))
print(colored(f'{Vardf2}','blue'))


def DataRelation(col,col1):
        dfcol= df1[col].dropna()
            #      print(dfcol)
        dfcol1= df2[col1].dropna()
            #      print(dfcol1
        intersec=pd.Series(np.intersect1d(dfcol, dfcol1))
            #      print(intersec)
        if not df1[col].empty and not df2[col1].empty:
            uque=dfcol.unique()
                #      print(uque)
            uque1=dfcol1.unique()
                #      print(uque1)
            
            DataMax = max(len(uque),len(uque1))
        #              print(DataMax)
            com_len=len(intersec)
                #      print(com_len)
            if DataMax !=0:
                percentage = str(int((100/DataMax)*com_len))+'%'
                #          print(percentage)
                if percentage != '0%':
                    print(colored(f'{col}: FROM DATA FRAME1 AND  {col1} FROM DATA FRAME2 MATCHED BY {percentage}','magenta'))
                    intrdf= pd.DataFrame(intersec, columns=[col])
                #              print(intrdf)
                    MrgData= len(intrdf.merge(df1[col],how='inner'))
                #              print(MrgData)
            
                    intrdf= pd.DataFrame(intersec, columns=[col1])
                #              print(intrdf)
                    MrgData1= len(intrdf.merge(df2[col1],how='inner'))
                #              print(MrgData1)
            
                    if MrgData==com_len:
                        if MrgData1==com_len:
                            print(colored('One to One','blue'))
                        else:
                            print(colored('One to Many','blue'))
                    else:
                        if MrgData1==com_len:
                            print(colored('Many to One','blue'))
                        else:
                            print(colored('Many to Many','blue')) 
                            
print(colored("*******************************************ANALYSING INTEGER TYPE COLUMN WITH PERCENTAGE AND TYPE OF JOIN********************\n",'green'))
for col in Intdf1:
    for col1 in Intdf2:
#         print(col)
#         print(col1)
        DataRelation(col,col1)
    
print(colored("********************************************ANALYSING VARCHAR TYPE COLUMN WITH PERCENTAGE AND TYPE OF JOIN*******************\n",'green'))            
for col in Vardf1:
    for col1 in Vardf2: 
#         print(col)
#         print(col1)
        DataRelation(col,col1)


print(colored("*******************************************ANALYSING DATE&TIME TYPE COLUMN WITH PERCENTAGE AND TYPE OF JOIN******************\n",'green'))            
for col in Datedf1:
#     print(col)
    for col1 in Datedf2:
#         print(col1) 
        DataRelation(col,col1)

        
```
